#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
month_encoded = pd.pivot_table(fitbit, index='Months_encoded', aggfunc=np.median)[['Steps']]
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(12,4))
ax.set_title('Steps by Month')
ax.plot(month_encoded.index, month_encoded['Steps'], linewidth=3, color='orange')
plt.show();

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
work_or_weekend_encoded = pd.pivot_table(fitbit, index='Work_or_Weekend', aggfunc=np.median)[['Steps']]
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(12,4))
ax.set_title('Steps by work day or weekend')
ax.bar(work_or_weekend_encoded.index, work_or_weekend_encoded['Steps'], linewidth=3, color='purple')
plt.xticks([0,1], ['Weekend','Work day'])
plt.show();
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
scatter_list = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

fig, ax = plt.subplots(ncols=1, nrows=3, figsize=(12,12))
for i in range(len(scatter_list)):
    title = scatter_list[i] + ' VS Steps'
    ax[i].set_title(title)
    ax[i].set_xlabel('Steps')
    ax[i].set_ylabel(scatter_list[i])
    ax[i].scatter(fitbit[scatter_list[i]], fitbit['Steps'], color='orange')
    fig.tight_layout()

plt.show();
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# your code here
mean_age = round(titanic['Age'].mean(),2)
median_age = round(titanic['Age'].median(),2)
mode_age = round(titanic['Age'].mode().values[0],2)


fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(12,8))
ax = sns.histplot(data=titanic, x=titanic['Age'],binwidth=10)
ax.axvline(mean_age, color='red', ls='-', lw=2, label=f'mean ({mean_age})')
ax.axvline(median_age, color='black', ls = '--', lw=2, label=f'median ({median_age})' )
ax.axvline(mode_age, color='green', ls='-.',lw=3, label=f'mode ({mode_age})')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,8))
ax= sns.boxplot(y=titanic['Age'])
plt.ylim(-2,85)
plt.show()
```

```python
"""
I think that there are two different ways to plot the statistics of the 'Age' column:
One is based on the histogram plot but adding the three central tendency measures: mean, median, mode. 
The other is based on the median and interquartile range, that is the boxplot (or box and whiskers).
Here we can see:
- Median: the horizontal line that is inside the blue box
- Third quartile (or 75 percentile): the upper side of the blue box
- First quartile (or 25 percentile): the lower side of the blue box
*Interquartile range = Q3 - Q1 
- Slightly atipical values (vertical lines)
- Extreme atipical values (points)
"""
```
